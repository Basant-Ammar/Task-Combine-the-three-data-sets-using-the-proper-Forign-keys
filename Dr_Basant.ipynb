```python
file_path = "Merge Two Data Sets Using Python (1).xlsx"
xls = pd.ExcelFile(file_path)
print(xls.sheet_names)
```

```python

sales_df = pd.read_excel(file_path, sheet_name="Sales")
product_df = pd.read_excel(file_path, sheet_name="Product")
salesperson_df = pd.read_excel(file_path, sheet_name="Salesperson")

print("Sales Data:\n", sales_df.head())
print("\nProduct Data:\n", product_df.head())
print("\nSalesperson Data:\n", salesperson_df.head())
```

```python

merged_df = sales_df.merge(product_df, on="ProductKey", how="left")


final_df = merged_df.merge(salesperson_df, on="EmployeeKey", how="left")

final_df.head()
```

```python
output_path = "/content/Combined_Data.xlsx"
final_df.to_excel(output_path, index=False)

files.download(output_path)
```
